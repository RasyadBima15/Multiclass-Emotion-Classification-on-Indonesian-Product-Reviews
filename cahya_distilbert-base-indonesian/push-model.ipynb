```python
import os
from huggingface_hub import login
from transformers import TFAutoModelForSequenceClassification, AutoTokenizer
```

```python
print("Silakan masukkan token Hugging Face Anda:")
login(token=os.getenv("HF_TOKEN"))
print("Login berhasil!")
```

```python
path_folder_lokal = "./model_distilbert_clickbait"
repo_id = "Rasyy/distilbert_indonesian_news_clickbait_detection"
```

```python
print(f"Memuat model dari {path_folder_lokal}...")
try:
    model = TFAutoModelForSequenceClassification.from_pretrained(path_folder_lokal)
    
    tokenizer = AutoTokenizer.from_pretrained(path_folder_lokal)
    
    print("Model dan tokenizer berhasil dimuat.")

except Exception as e:
    print(f"Gagal memuat model. Pastikan path sudah benar dan file lengkap.")
    print(f"Error: {e}")
    exit()
```

```python
print(f"Mengunggah file ke repositori: {repo_id}...")

try:
    model.push_to_hub(repo_id)
    print("File model berhasil diunggah.")

    tokenizer.push_to_hub(repo_id)
    print("File tokenizer berhasil diunggah.")

    print("\n--- Selesai! ---")
    print(f"Model Anda sekarang tersedia di:")
    print(f"https://huggingface.co/{repo_id}")

except Exception as e:
    print(f"Terjadi kesalahan saat mengunggah: {e}")
```
